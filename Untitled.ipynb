### anime recommender system using clustering | collaborative based

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
anime = pd.read_csv('data/animes.csv')
anime
```

```python
anime.shape
```

```python
anime.columns
```

```python
anime = anime[['uid', 'title', 'synopsis', 'genre','img_url']]
```

```python
anime.rename(columns={'uid': 'anime_uid'}, inplace=True)
anime
```

```python
users = pd.read_csv('data/profiles.csv')
users
```

```python
users = users[['profile', 'favorites_anime']]
users
```

```python
user_anime = pd.read_csv('data/reviews.csv')
user_anime
```

```python
user_anime = user_anime[['profile','anime_uid','score','scores']]
user_anime
```

```python
searched_anime = anime[anime['anime_uid'] == 34096]
searched_anime
```

```python
anime.drop_duplicates(subset=['title'], inplace=True)
```

```python
anime.shape
```

```python
user_anime['profile'].value_counts()
```

```python
mask = user_anime['profile'].value_counts()>50
print(mask.sum())
```

```python
#making sure the uid is NOT the users id

user_anime[user_anime['profile'] == 'DesolatePsyche']
```

```python
y = mask[mask].index
y
```

```python
user_anime[user_anime['profile'].isin(y)]
```

```python
ratings = user_anime[user_anime['profile'].isin(y)]
```

```python
ratings.head()
```

```python
ratings[ratings['anime_uid']== 28891]
```

```python
anime_ratings = ratings.merge(anime, on= 'anime_uid')
anime_ratings
```

```python
#how many times these animes got rated 

num_rating = anime_ratings.groupby('title')['score'].count().reset_index()
```

```python
num_rating
```

```python
# some of them got rated only very few times; they are bad

num_rating.rename(columns={'score':'number_of_ratings'}, inplace=True)
```

```python
num_rating.head()
```

```python
final_rating = anime_ratings.merge(num_rating, on='title')
final_rating
```

```python
final_rating['number_of_ratings'].max()
```

```python
percentile_75 = final_rating['number_of_ratings'].quantile(0.75)
print(percentile_75)
```

```python
final_rating = final_rating[final_rating['number_of_ratings']>=60]
```

```python
final_rating
```

```python
final_rating.drop_duplicates(['title','profile'], inplace= True)
```

```python
final_rating
```

```python
anime_pivot = final_rating.pivot_table(columns='profile', index='title', values='score')
```

```python
anime_pivot
```

```python
anime_pivot.shape
```

```python
anime_pivot.fillna(0, inplace = True)
```

```python
anime_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
anime_sparse = csr_matrix(anime_pivot)
```

```python
anime_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(anime_sparse)
```

```python
distance, suggestion = model.kneighbors(anime_pivot.iloc[23,:].values.reshape(1,-1), n_neighbors=5)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(anime_pivot.index[suggestion[i]])
```

```python
import pickle
pickle.dump(model, open('artifacts/model.pkl', 'wb'))
pickle.dump(anime_name, open('artifacts/anime_name.pkl', 'wb'))
pickle.dump(final_rating, open('artifacts/final_rating.pkl', 'wb'))
pickle.dump(anime_pivot, open('artifacts/anime_pivot.pkl', 'wb'))
```

```python
def recommend_anime(anime_name):
    anime_id = np.where(anime_pivot.index == anime_name)[0][0]
    distance, suggestion = model.kneighbors(anime_pivot.iloc[anime_id,:].values.reshape(1,-1), n_neighbors=10)
    for i in range(len(suggestion)):
        animes = anime_pivot.index[suggestion[i]]
        for j in animes:
            print(j)
```

```python
book_name='Bleach'
recommend_anime(book_name)
```
